```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

I = mpimg.imread('./data13.bmp')
print(I.shape)
plt.imshow(I)
plt.show()

# transform RGB to gray value image

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
 
gray = rgb2gray(I)    
#  plt.imshow(gray, cmap = plt.get_cmap('gray'))
print(gray.shape)
plt.imshow(gray, cmap='Greys_r')
plt.axis('off')
```

```python
## transform RGB to gray value image
def RGB2Grey(I):
    height, width = I.shape[0], I.shape[1]
    gray_I = x = np.ones([height,width], dtype = float)
    for i in range(height):
        for j in range(width):
            [r,g,b] = I[i][j]
            gray_I[i][j] = 0.299*r +  0.587*g + 0.114*b
    return gray_I

plt.imshow(gray_I)
plt.imshow(gray, cmap='Greys_r')
plt.show()
```
